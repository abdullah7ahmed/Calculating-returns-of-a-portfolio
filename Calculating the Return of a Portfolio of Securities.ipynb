Loading the data for a portfolio composed of 5 stocks - British Petroleum, Ford, Exxon, Lincoln, and Apple *(5_stocks_2000_2017.csv)*.

### Calculating the Return of a Portfolio of Securities

Obtaining the simple return of the securities in the portfolio and storing the results in a new table.

```python
returns = (mydata / mydata.shift(1)) -1
returns.head()
```

We will create an equally weighted portfolio with the name weights.

```python
weights = np.array([0.20, 0.20, 0.20, 0.20, 0.20])
```

Obtaining the annual returns of each of the stocks and then calculate the dot product of these returns and the weights.

```python
annual_returns = returns.mean() * 250
annual_returns.head()
```

```python
np.dot(weights, annual_returns)
```

Transforming the result into a percentage form. 

```python
portfolio_returns = str(round(np.dot(weights, annual_returns) * 100 )) + ' %'
print(portfolio_returns)
```

According to industry benchmark, 10% of more portfolio return is considered satisfactory.
